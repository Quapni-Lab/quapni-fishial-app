```python
import os
import cv2
os.environ['KMP_DUPLICATE_LIB_OK']='True'
from classification_task.inference_class import EmbeddingClassifier

classification_package = './classification_task/model'
classification_path = os.path.join(classification_package, 'model.ts')
data_base_path = os.path.join(classification_package, 'embeddings.pt')
data_idx_path = os.path.join(classification_package, 'idx.json')

device = 'cpu'
model_classifier = EmbeddingClassifier(
        classification_path,
        data_base_path,
        data_idx_path,
        device=device)

#your/image/path.jpeg
image_path = './example/result.jpg'
#read image which you want to classify
im = cv2.imread(image_path)
# im = cropped_image

#single image inference
single_output = model_classifier.inference_numpy(im) 

# you will get list of [['Tinca tinca', 1.0, ['008f26d7-ea82-5340-8902-6b0b20681892', '284620', 593428]]] where ['species_name',accuracy_between_0.0_and_1.0, Unnecessary info]

print(f"Single inference output: {single_output}")
```

```python
import os
import json

def read_json(path):
    if os.path.isfile(path):
        with open(path) as f:
            data = json.load(f)
        return data
    else:
        return None

labels_file = read_json(data_idx_path)
```

```python
# labels_file['categories']
names = [value['name'] for value in labels_file['categories'].values()]
for i in range(len(names)):
    print(names[i])
```
